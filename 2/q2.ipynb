```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def show(img):
    cv2.imshow("image",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
```

```python
def error(img1,img2,m,n):
    sum  = 0
    h = img1.shape[0]
    for i in range(h):
        for j in range(m,n):
            sum += ((img1[i,j] - img2[i,j])*(img1[i,j] - img2[i,j]))
    return sum/((n-m)*h)
```
