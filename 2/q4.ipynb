```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def show(img):
    cv2.imshow("image",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
img  = cv2.imread("data/Building.jpg",0)
show(img)
```

```python
out = cv2.Canny(img,100,150) 
plt.imshow(out,cmap='gray')
```

```python
blur = cv2.GaussianBlur(img,(5,5),0) # if sigmax is 0/specified then sigmay is sigmax, and if sigma 
plt.imshow(blur, cmap='gray')
plt.show()
```

```python
log = cv2.Laplacian(blur, cv2.CV_8U)
plt.imshow(log, cmap='gray')
plt.show()
log_8u = log/log.max()
print(log_8u.max())
log_8u = log_8u*255
log_8u = log_8u.astype(np.uint8)
show(log_8u)
print(log_8u.dtype)
```
