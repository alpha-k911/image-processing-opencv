```python
def generateHistogram(image):
    data = {}
    for i in range(256):
        data[i] = 0
    h,w,_ = image.shape
    for i in range(w):
        for j in range(h):
            data[image[j][i][0]] += 1
            
    x_range = np.arange(0,256,1)
    plt.xlim([0,256])
    plt.bar(x_range,data.values())
#     return data
```
